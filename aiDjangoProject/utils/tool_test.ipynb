````python
import re
import os

def extract_structural_code(text):
    """
    提取结构化编程风格参考下的C代码
    """
    pattern = re.compile(
        r'###\s*结构化编程风格参考\s*###\s*```c(.*?)```', re.DOTALL)
    match = pattern.search(text)
    return match.group(1).strip() if match else ""

def extract_modular_code(text):
    """
    提取模块化编程风格参考下的C代码
    """
    pattern = re.compile(
        r'###\s*模块化编程风格参考\s*###\s*```c(.*?)```', re.DOTALL)
    match = pattern.search(text)
    return match.group(1).strip() if match else ""

#def extract_procedural_code(text):
    """
    提取面向过程编程风格参考下的C代码
    """
   # pattern = re.compile(
    #    r'###\s*面向过程(编程)?风格参考\s*###\s*```c(.*?)```', re.DOTALL)
   # match = pattern.search(text)
   # return match.group(1).strip() if match else ""

def extract_procedural_code(text):
    """
    提取面向过程编程风格参考下的C代码
    """
    pattern = re.compile(
        r'###\s*面向过程(编程)?风格参考\s*###\s*```c(.*?)```', re.DOTALL)
    match = pattern.search(text)
    return match.group(2).strip() if match else ""

def save_code(code, filename):
    """
    保存代码到 ./C51_codes/ 下指定文件
    """
    os.makedirs("./C51_codes/", exist_ok=True)
    with open(f"./C51_codes/{filename}", "w", encoding="utf-8") as f:
        f.write(code)
    print(f"{filename} 已保存到 ./C51_codes/{filename}")

# 示例主流程
def extract_and_save_all(text):
    structural_code = extract_structural_code(text)
    modular_code = extract_modular_code(text)
    procedural_code = extract_procedural_code(text)
    if structural_code:
        save_code(structural_code, "structural.c")
    else:
        print("未找到结构化编程风格代码")
    if modular_code:
        save_code(modular_code, "modular.c")
    else:
        print("未找到模块化编程风格代码")
    if procedural_code:
        save_code(procedural_code, "procedural.c")
    else:
        print("未找到面向过程编程风格代码")
````
